```python
fiber = telecom.loc[lambda df: telecom['InternetService'] == 'Fiber optic', :]
```

```python
sns.set(style = "darkgrid")

c = sns.factorplot(x = "Contract", y="MonthlyCharges", hue="Churn", data = fiber, size = 6, kind = "bar", palette ="pastel",)
c.despine(left=True)
c.set_ylabels("Monthly charges in $")
plt.title('Barplot: Monthly charges for fiber optic users')
```
